```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path_1 = "test_jpeg.csv"  
csv_path_2 = "test_heic.csv"  
```

```python

df1 = pd.read_csv(csv_path_1)
df2 = pd.read_csv(csv_path_2)

# --- Plot -------------------------------------------------------------------
plt.figure(figsize=(7, 4))

plt.plot(df1["quality"], df1["psnr"], marker="o",
         linewidth=2, label="JPEG")
plt.plot(df2["quality"], df2["psnr"], marker="s",
         linewidth=2, label="HEIC")

plt.xlabel("Quality")
plt.ylabel("PSNR")
plt.title("PSNR vs. Quality")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```
